```python
car_sales=pd.read_csv('car-sales.csv')
car_sales.head()
```

```python
car_sales.info()
```

```python
plt.plot(car_sales['Price'])
```

```python
plt.plot(car_sales['Odometer( KM)'])
```

##### The plot() function works for Price beacuse it is an object, whereas Odometer (KM) is an int, so the plot() function won't work

```python
car_sales.columns
```

```python
car_sales["Price"] = car_sales["Price"].replace("[$,]", "", regex=True).astype(int)
car_sales.head()
```

```python
car_sales['Total Sales'] = car_sales['Price'].cumsum()
car_sales.info()
```

```python
car_sales['Sales Date'] = pd.date_range(start ='2-11-2023', periods = len(car_sales))
car_sales.info()
```

```python
car_sales.head()
```

```python
price_total_sales=plt.plot(car_sales['Price'], car_sales['Total Sales'])

plt.title('Price vs total sales line plot')

plt.xlabel('Price')
plt.ylabel('Total Sales')

plt.show()
```

```python
odometer_price=plt.scatter(car_sales['Odometer (KM)'], car_sales['Price'])
plt.suptitle('Odometer vs Price scatter plot')
plt.xlabel('Odometer (KM)')
plt.ylabel('Price')
plt.show()
```

```python
make_odometer=plt.bar(car_sales['Make'], car_sales['Odometer (KM)'])
plt.suptitle('Make vs Odometer bar plot')
plt.xlabel('Make')
plt.ylabel('Odometer (KM)')
plt.show()
```

```python
make_hist=plt.hist(car_sales['Make'])
plt.suptitle('Histogram of car make')
plt.xlabel('Make')
plt.show()
```

```python
fig=plt.figure(figsize=(15, 10))

ax1=fig.add_subplot(221)
ax1.set_title('Price vs total sales line plot')

ax1.plot(car_sales['Price'], 
        car_sales['Total Sales'])

ax2=fig.add_subplot(222)
ax2.set_title('Odometer vs Price scatter plot')

ax2.scatter(car_sales['Odometer (KM)'], 
          car_sales['Price'])

ax3=fig.add_subplot(223)
ax3.set_title('Make vs Odometer bar plot')

ax3.bar(car_sales['Make'], 
         car_sales['Odometer (KM)'])

ax4=fig.add_subplot(224)
ax4.set_title('Histogram of car make')

ax4.hist(car_sales['Make'])

plt.show()
```

```python
box_plot=sns.boxplot(x='Colour', y='Total Sales', data=car_sales)
```

```python
car_sales['Total Sales'].corr(car_sales['Odometer (KM)'])
```

```python
lm_plot=sns.lmplot(x='Total Sales',
                  y='Odometer (KM)',
                  data=car_sales,
                  hue='Make',
                  ci=False)
```

```python
pair_plot=sns.pairplot(car_sales)
```

```python
heat_map_corr=car_sales[['Total Sales',
                        'Price',
                        'Odometer (KM)']].corr(method='pearson')
print(heat_map_corr)
cols=['Total Sales', 'Price', 'Odometer (KM)']

heat_map=sns.heatmap(heat_map_corr,
                     annot=True,
                    xticklabels=cols,
                    yticklabels=cols)
```

#### The correlation is either very low or negative, meaning Total Sales, Price, and Odometer (KM) are either not related or inversely related
